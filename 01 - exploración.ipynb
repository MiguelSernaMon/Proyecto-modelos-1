## load `train.csv`

```python
import pandas as pd
import numpy as np

z = pd.read_csv("train.csv")
print ("shape of loaded dataframe", z.shape)
```

```python
z.head()
```

#**Test #1: ESTUDIANTES DE ENFERMERIA, FILTRADO POR ESTRATO MAYOR A 3**

```python


z['FAMI_ESTRATOVIVIENDA'] = z['FAMI_ESTRATOVIVIENDA'].replace('Sin Estrato', np.nan)

z = z.dropna(subset=['FAMI_ESTRATOVIVIENDA'])

z['FAMI_ESTRATOVIVIENDA'] = z['FAMI_ESTRATOVIVIENDA'].str.replace('Estrato ', '').astype(int)

filtro = (z['ESTU_PRGM_ACADEMICO'] == 'ENFERMERIA') & (z['FAMI_ESTRATOVIVIENDA'] > 3)
z_filtrado = z[filtro]
z_filtrado.head()

```

#**Test #2: ESTUDIANTES DE SANTANDER, FILTRADO POR FAMILIA NO TIENE INTERNET**


```python


filtro = (z['ESTU_PRGM_DEPARTAMENTO'] == 'SANTANDER') & (z['FAMI_TIENEINTERNET'] == 'No')
z_filtrado = z[filtro]
z_filtrado.head()
```

#**Test #3: PORCENTAJE DE ESTUDIANTES DE ANTIOQUIA CON PROGRAMA CIENCIA POLITICA


```python
numberOfPoliticalStudentsOfAntioquia = z[(z['ESTU_PRGM_DEPARTAMENTO'] == 'ANTIOQUIA') & (z['ESTU_PRGM_ACADEMICO'] == "CIENCIA POLITICA")].shape[0]

print("El porcentaje de Esudiantes de antioquia que estudia ciencias politicas es:")
print((numberOfPoliticalStudentsOfAntioquia/z.shape[0])*100,"%")



```
